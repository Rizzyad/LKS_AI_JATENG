```python
import pathlib
import numpy as np

dir_path = pathlib.Path("subset_datasetlomba")
class_names = np.array(sorted([item.name for item in dir_path.glob('*')]))

print(class_names)
```

```python
import random
import matplotlib.pyplot as plt
import matplotlib.image as pltimg

def view_random_image(target_dir, target_class):
  target_folder = os.path.join(target_dir, target_class)
  random_image = random.choice(os.listdir(target_folder))

  img_path = os.path.join(target_folder, random_image)
  img = pltimg.imread(img_path)

  plt.imshow(img)
  plt.title(target_class)
  plt.axis(False)

  print(f"Image Shape : {img.shape}")

  return img
```

```python
img_dict = {}

for target_class in class_names:
  img = view_random_image("subset_datasetlomba/", target_class)
  img_dict[target_class] = img

  plt.show()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen_augmented = ImageDataGenerator(rescale=1/255.,
                                             rotation_range=15,
                                             width_shift_range=0.1,
                                             height_shift_range=0.1,
                                             shear_range=0.1,
                                             zoom_range=0.1,
                                             horizontal_flip=True,
                                             fill_mode='nearest',
                                             validation_split=0.2)

train_datagen = ImageDataGenerator(rescale=1/255.,
                                   validation_split=0.2)

valid_datagen = ImageDataGenerator(rescale=1/255.,
                                   validation_split=0.2)
```

```python
train_data_augmented = train_datagen_augmented.flow_from_directory("subset_datasetlomba",
                                                                   batch_size=32,
                                                                   target_size=(224,224),
                                                                   class_mode='binary',
                                                                   seed=42,
                                                                   shuffle=False,
                                                                   subset='training')

train_data = train_datagen.flow_from_directory("subset_datasetlomba",
                                                                   batch_size=32,
                                                                   target_size=(224,224),
                                                                   class_mode='binary',
                                                                   seed=42,
                                                                   shuffle=False,
                                                                   subset='training')

valid_data = valid_datagen.flow_from_directory("subset_datasetlomba",
                                                                   batch_size=32,
                                                                   target_size=(224,224),
                                                                   class_mode='binary',
                                                                   seed=42,
                                                                   shuffle=False,
                                                                   subset='validation')
```

```python
train_data_augmented_shuffled = train_datagen_augmented.flow_from_directory("subset_datasetlomba",
                                                                   batch_size=32,
                                                                   target_size=(224,224),
                                                                   class_mode='binary',
                                                                   seed=42,
                                                                   shuffle=True,
                                                                   subset='training')
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, BatchNormalization, Dropout, Flatten, Dense, GlobalMaxPool2D
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
```

```python
checkpoint = ModelCheckpoint("best_weight.h5", verbose=1, save_best_only=True, save_weight_only=True)
early_stop = EarlyStopping(patience=4)
```

```python
model_1 = Sequential([
    Conv2D(32, 3, activation='relu', input_shape=(224,224,3)),
    MaxPool2D(2),
    Conv2D(64, 3, activation='relu'),
    MaxPool2D(2),
    Conv2D(128, 3, activation='relu'),
    MaxPool2D(2),
    Flatten(),
    Dense(1, activation='sigmoid')
])

model_1.compile(
    loss = 'binary_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)

history_1 = model_1.fit(train_data_augmented_shuffled,
                        epochs=5,
                        steps_per_epoch=len(train_data_augmented_shuffled),
                        validation_data=valid_data,
                        validation_steps=len(valid_data),
                        callbacks=[checkpoint, early_stop],
                        verbose=True)
```

```python
def plot_loss_curves(history):
  loss = history.history['loss']
  val_loss = history.history['val_loss']

  accuracy = history.history['accuracy']
  val_accuracy = history.history['val_accuracy']

  epochs = range(len(history.history['loss']))

  plt.plot(epochs, loss, label='training_loss')
  plt.plot(epochs, val_loss, label='validation_loss')
  plt.title('Loss')
  plt.xlabel('Epoch')
  plt.legend()

  plt.figure()
  plt.plot(epochs, accuracy, label='training_accuracy')
  plt.plot(epochs, val_accuracy, label='validation_accuracy')
  plt.title('Accuracy')
  plt.xlabel('Epoch')
  plt.legend()
```

```python
plot_loss_curves(history_1)
```

```python
import gradio as gr

num_classes = train_data_augmented_shuffled.num_classes
num_classes
```

```python
# Binary clssification
model = tf.keras.models.load_model('model_1.h5')

def classify_image(inp):
  inp = tf.image.resize(inp, size=[224,224])
  inp = tf.expand_dims(inp, axis=0)
  inp = inp/255.

  prediction = model.predict(inp)
  prediction = np.round(prediction.flatten(), decimals = 2)
  print(prediction)
  confidances = {class_names[i]: (1-float(prediction)) if i == 0 else float(prediction) for i in range(len(class_names))}
  return confidances

gr.Interface(theme=gr.themes.Soft(),
             title = 'Cat Vs Dog Image Classification',
             description='<center>Klassifikasi Gambar Kucing Atau Anjing Menggunakan Tensorflow Dan Gradio</center>',
             article = 'Input gambar menggunakan webcam lalu klik submit untuk membuat prediksi gambar',
             fn = classify_image,
             inputs = gr.Image(source='webcam', input_shape=(224,224), label='Input gambar'),
             outputs = gr.Label(num_top_classes=num_classes)).launch(share=True, debug=True)
```
